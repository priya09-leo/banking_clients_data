## Create income bands

### Subtask:
Define income bands based on the 'Estimated Income' column and create a new column 'Income band' in the DataFrame.


**Reasoning**:
Define income bands based on the 'Estimated Income' column and create a new column 'Income band' in the DataFrame using `pd.cut()`.



```python
bins = [df['Estimated Income'].min(), 50000, 150000, 300000, df['Estimated Income'].max()]
labels = ['Low', 'Medium', 'High', 'Very High']
df['Income band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels, include_lowest=True)
display(df.head())
```

## Heatmaps

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Clean column names
df.columns = df.columns.str.strip()

# Updated column list (no extra spaces)
numerical_cols = [
    'Estimated Income', 'Superannuation Savings', 'Credit Card Balance',
    'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts',
    'Foreign Currency Account', 'Business Lending',
    'Properties Owned', 'Risk Weighting'
]

# Correlation matrix
correlation_matrix = df[numerical_cols].corr()

# Plot heatmap
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix")
plt.show()
```

## Insights of EDA

1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts", "Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain high balances in one account as well.
